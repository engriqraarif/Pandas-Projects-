```python
import pandas as pd 
```

```python
import seaborn as sns
```

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Customer Withdraw.csv")
```

```python
df.info()
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ",0)
df["TotalCharges"]=df["TotalCharges"].astype(float)
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum().sum()
```

```python
df.info()
```

```python
df = df.rename(columns={"Churn" : "Withdraw"})

```

```python
df["SeniorCitizen"] = df["SeniorCitizen"].replace({1: "Yes", 0: "No"})
```

```python
df.info()
```

```python
df.head()
```

```python
Withdraw_count_plot = sns.countplot(x= "Withdraw",data=df)
Withdraw_count_plot.bar_label(Withdraw_count_plot.containers[0])
plt.title("Bar chart of withdrawl")
```

```python
Withdraw_in_pie = df.groupby("Withdraw").agg({"Withdraw":"count"})
plt.pie(Withdraw_in_pie["Withdraw"],labels=Withdraw_in_pie.index,autopct="%1.3f%%")
plt.title("It is a pie Chart of withdraw customer in percentage",fontsize=10,color="red")
plt.show()
```

```python
gender_count=sns.countplot (x="gender",data=df,hue="Withdraw")
for container in gender_count.containers:gender_count.bar_label(container,fmt="%d",padding=)
```

```python
Citizen_count_plot = sns.countplot(x= "SeniorCitizen",data=df,palette="Set2")
Citizen_count_plot.bar_label(Citizen_count_plot.containers[0])
plt.title("Bar chart of senior citizen")
```

```python
plt.figure(figsize=(9,3))
tenure_count=sns.histplot(x="tenure",data=df,hue="Withdraw",bins=72,palette="Set2")
```

```python
withdraw=sns.countplot(x="Contract",data=df,hue="Withdraw",palette="Set2")
for c in withdraw.containers:withdraw.bar_label(c)
plt.title("It i a bar hart of withdrawcustomer by contract")
```

```python
plt.figure(figsize=(10,8 ))
columns=["MultipleLines" ,"InternetService","OnlineSecurity","OnlineBackup","DeviceProtection",
         "TechSupport","StreamingTV", "StreamingMovies"]
n_cols=3
n_rows=(len(columns)+n_cols-1)//n_cols
fig,axes=plt.subplots(n_rows,n_cols,figsize=(15,n_rows+6),constrained_layout=True)
axes=axes.flatten()
for i, col in enumerate(columns):
    sns.countplot(x=col,hue=col,data=df,ax=axes[i],palette="Set2",legend=False)
    axes[i].set_title(f'count plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("count")
    axes[i].tick_params(axis="x",rotation=45)
    for container in axes[i].containers:axes[i].bar_label(container,padding=3,fmt="%d")
    axes[i].set_ylim(0,axes[i].get_ylim()[1]*1.3)

for j in range(i+1,len(axes)):
     fig.delaxes(axes[j])
fig.subplots_adjust(top=0.92,hspace=0.5,wspace=0.4)
```
